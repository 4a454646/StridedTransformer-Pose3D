```python
# annotating frames on video

video_path = "/home/jeff/Documents/Code/StridedTransformer-Pose3D/demo/video/daniel_test_scaled.mp4"
import cv2
cap = cv2.VideoCapture(video_path)
frame_number = 0
with_frame = []
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        # write frame_number onto the frame
        out_frame = cv2.putText(frame, str(frame_number), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        with_frame.append(out_frame)
        frame_number += 1
    else:
        break
cap.release()
cv2.destroyAllWindows()
print(with_frame[0].shape[:-1])
# create a video out of the list of frames from with_frame
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('frame_annotated.mp4', fourcc, 30, with_frame[0].shape[:-1][::-1])
for i in range(len(with_frame)):
    # print("writing")
    out.write(with_frame[i])
out.release()
```
